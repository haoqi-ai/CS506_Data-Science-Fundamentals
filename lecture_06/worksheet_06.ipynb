### Kmeans ++

a) What is the difference between K means and K means ++?

They are different in the initialization method for the cluster centers. K-means typically starts with random initialization, which can sometimes lead to suboptimal clustering. K-means++, on the other hand, improves by selecting initial centers except the first one with a probability proportional to their squared distance from the nearest existing center. 

b) What are some limitations of K means ++?

- If the natural clusters within data vary significantly in size, k-means++ may still struggle to identify smaller clusters accurately. 

- K-means++ may perform poorly with datasets where clusters have widely differing variances or densities. Since the distance metric (typically Euclidean) does not account for density, dense clusters can be merged, and sparse clusters can be split. 

- K-means++ is designed to identify spherical clusters due to the use of Euclidean distance. Therefore, it may not identify clusters correctly if the clusters have non-spherical shapes, such as elongated clusters or clusters with complex geometries.

### Input:

### Task:
- Implement a hierarchical clustering algorithm using Hamming distance as the metric clustering DNA sequences.

### Definition of Hamming Distance:

The Hamming distance between two strings of equal length is the number of positions at which the corresponding symbols are different. Mathematically, if we have two strings, $s$ and $t$, of equal length, then the Hamming distance $H(s, t)$ is given by:

$$ H(s, t) = \sum_{i=1}^{n} [s_i \neq t_i] $$

where $n$ is the length of the strings, $s_i$ and $t_i$ are the characters at position $i$ in $s$ and $t$ respectively, and $[s_i \neq t_i]$ is an indicator function, equal to 1 if $s_i \neq t_i$ and 0 otherwise.


### Guidelines:
1. **Read the Dataset**: Choose an appropriate data structure.
2. **Compute Hamming Distance**: Implement a function to calculate the Hamming distance between any two sequences.
3. **Hierarchical Clustering**: Apply the hierarchical clustering algorithm using the single-linkage method.
4. **Dendrogram**: Generate a dendrogram to visualize the clustering.
5. **NOTE**: You may use any Python library, but be sure to understand the underlying algorithm.

```python
sequences = [
 'ACGTGGTCTTAA',
 'ACGTCGTCTTAC',
 'ACGTGGTCTTAC',
 'ACGTAGTCTTAA',
 'ACGTGGTCTTCC',
 'ACGTGGTCTTAG',
 'CTGTTAAATAAG',
 'GGTTAGAACACG',
 'AGTGGTTGAAGT',
 'GGCTTACACCCT',
 'AGATTGTCCACT',
 'CATGCGGTCAAC',
 'ATATATCATAGC',
 'TTTGCGGTTGGA',
 'GAATGGTCAGAA',
 'GTGATGCTGTCT']
```

```python
def cal_hamming_dis(str1, str2):
    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length!")
    return sum(c1 != c2 for c1, c2 in zip(str1, str2))
```

```python
import numpy as np 
import matplotlib.pyplot as plt
# https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html
from scipy.cluster.hierarchy import linkage, dendrogram
import warnings
warnings.filterwarnings('ignore')  

n = len(sequences)
matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        matrix[i][j] = cal_hamming_dis(sequences[i], sequences[j])

cluster = linkage(matrix, method='single')
print(cluster.shape)
print(cluster[:5])
```

```python
plt.figure(figsize=(10, 8))
dendrogram(cluster, labels=sequences, leaf_rotation=45, leaf_font_size=8)
plt.xlabel('DNA Sequence', fontsize=12)
plt.ylabel('Hamming Distance', fontsize=12)
plt.show()
```
